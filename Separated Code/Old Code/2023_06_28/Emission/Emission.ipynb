## Import Necessary data

```python
# Import FA data
FA_data = pd.read_csv('/Users/akihi/Library/CloudStorage/OneDrive-PrincetonUniversity/Projects/Building Material in Cities/1st Research/Code/Separated Code/Floor Area/Merged.csv')
FA_data = FA_data.set_index(FA_data.iloc[:, 0])
FA_data
```

```python
# Emission Factor
EF_data = pd.read_csv('/Users/akihi/Library/CloudStorage/OneDrive-PrincetonUniversity/Projects/Building Material in Cities/1st Research/Code/Separated Code/Input/Emission Factor.csv')

EF_total_LF = EF_data['LF'][0]
EF_total_RC = EF_data['RC'][0]
EF_total_MT = EF_data['MT'][0]
EF_total_CSA = EF_data['CSA'][0]
EF_total_AAC = EF_data['AAC'][0]

EF_production_LF = EF_data['LF'][1]
EF_production_RC = EF_data['RC'][1]
EF_production_MT = EF_data['MT'][1]
EF_production_CSA = EF_data['CSA'][1]
EF_production_AAC = EF_data['AAC'][1]

EF_storage_LF = EF_data['LF'][2]
EF_storage_RC = EF_data['RC'][2]
EF_storage_MT = EF_data['MT'][2]
EF_storage_CSA = EF_data['CSA'][2]
EF_storage_AAC = EF_data['AAC'][2]
```

## Estimate Emission

### Production Emission in 2 or 22 years

```python
# Prepare the dataframe
column = {'BAU2020_Total': [], 'BAU2020_PE': [], 'BAU2020_CS': [],
'BAU2040_Total': [], 'BAU2040_PE': [], 'BAU2040_CS': [],
'Compact_City_Total': [], 'Compact_City_PE': [], 'Compact_City_CS': [],
'Effective_Design_Total': [], 'Effective_Design_PE': [], 'Effective_Design_CS': [],
'Intensive_Use_Total': [], 'Intensive_Use_PE': [], 'Intensive_Use_CS': [],
'Mass_Timber_Total': [], 'Mass_Timber_PE': [], 'Mass_Timber_CS': [],
'Alternative_Cement_Total': [], 'Alternative_Cement_PE': [], 'Alternative_Cement_CS': [],
"'+ Effective_Design_Total": [], "'+ Effective_Design_PE": [], "'+ Effective_Design_CS": [],
"'+ Intensive_Use_Total": [], "'+ Intensive_Use_PE": [], "'+ Intensive_Use_CS": [],
"'+ Mass_Timber_Total": [], "'+ Mass_Timber_PE": [], "'+ Mass_Timber_CS": [],
"'+ Alternative_Cement_Total": [], "'+ Alternative_Cement_PE": [], "'+ Alternative_Cement_CS": []}
index = ['LF', 'RC', 'MT', 'CSA cement', 'AAC',
         'Total (with CSA)', 'Total (with AAC)']

Emission_data = pd.DataFrame(index=index, columns=column.keys())
Emission_data = Emission_data.fillna(0)
```

```python
# Calculate emission
list1 = ['BAU2020_PE', 'BAU2040_PE', 'Compact_City_PE', 'Effective_Design_PE', 'Intensive_Use_PE', 'Mass_Timber_PE', 'Alternative_Cement_PE', "'+ Effective_Design_PE", "'+ Intensive_Use_PE", "'+ Mass_Timber_PE", "'+ Alternative_Cement_PE"]
list2 = ['BAU 2020', 'BAU 2040','Compact City', 'Effective Design', 'Intensive Use', 'Mass-Timber', 'Alternative Cement', "'+ Effective Design", "'+ Intensive Use", "'+ Mass-Timber", "'+ Alternative Cement"]

for scenario, sub_scenario in zip(list1, list2):
    Emission_data.loc['LF', scenario] = FA_data.loc[sub_scenario, 'New_FA_LF'] * EF_production_LF
    Emission_data.loc['RC', scenario] = FA_data.loc[sub_scenario, 'New_FA_RC'] * EF_production_RC
    Emission_data.loc['MT', scenario] = FA_data.loc[sub_scenario, 'New_FA_MT'] * EF_production_MT
    Emission_data.loc['CSA cement', scenario] = FA_data.loc[sub_scenario, 'New_FA_CSA'] * EF_production_CSA
    Emission_data.loc['AAC', scenario] = FA_data.loc[sub_scenario, 'New_FA_AAC'] * EF_production_AAC
```

```python
# Calculate carbon storage
list1 = ['BAU2020_CS', 'BAU2040_CS', 'Compact_City_CS', 'Effective_Design_CS', 'Intensive_Use_CS', 'Mass_Timber_CS', 'Alternative_Cement_CS', "'+ Effective_Design_CS", "'+ Intensive_Use_CS", "'+ Mass_Timber_CS", "'+ Alternative_Cement_CS"]
list2 = ['BAU 2020', 'BAU 2040','Compact City', 'Effective Design', 'Intensive Use', 'Mass-Timber', 'Alternative Cement', "'+ Effective Design", "'+ Intensive Use", "'+ Mass-Timber", "'+ Alternative Cement"]

for scenario, sub_scenario in zip(list1, list2):
    Emission_data.loc['LF', scenario] = FA_data.loc[sub_scenario, 'New_FA_LF'] * EF_storage_LF
    Emission_data.loc['RC', scenario] = FA_data.loc[sub_scenario, 'New_FA_RC'] * EF_storage_RC
    Emission_data.loc['MT', scenario] = FA_data.loc[sub_scenario, 'New_FA_MT'] * EF_storage_MT
    Emission_data.loc['CSA cement', scenario] = FA_data.loc[sub_scenario, 'New_FA_CSA'] * EF_storage_CSA
    Emission_data.loc['AAC', scenario] = FA_data.loc[sub_scenario, 'New_FA_AAC'] * EF_storage_AAC

Emission_data
```

```python
# Calculate total CO2 balance
list1 = ['BAU2020_Total', 'BAU2040_Total', 'Compact_City_Total', 'Effective_Design_Total', 'Intensive_Use_Total', 'Mass_Timber_Total', 'Alternative_Cement_Total', "'+ Effective_Design_Total", "'+ Intensive_Use_Total", "'+ Mass_Timber_Total", "'+ Alternative_Cement_Total"]
list2 = ['BAU 2020', 'BAU 2040','Compact City', 'Effective Design', 'Intensive Use', 'Mass-Timber', 'Alternative Cement', "'+ Effective Design", "'+ Intensive Use", "'+ Mass-Timber", "'+ Alternative Cement"]

for scenario, sub_scenario in zip(list1, list2):
    Emission_data.loc['LF', scenario] = FA_data.loc[sub_scenario, 'New_FA_LF'] * EF_total_LF
    Emission_data.loc['RC', scenario] = FA_data.loc[sub_scenario, 'New_FA_RC'] * EF_total_RC
    Emission_data.loc['MT', scenario] = FA_data.loc[sub_scenario, 'New_FA_MT'] * EF_total_MT
    Emission_data.loc['CSA cement', scenario] = FA_data.loc[sub_scenario, 'New_FA_CSA'] * EF_total_CSA
    Emission_data.loc['AAC', scenario] = FA_data.loc[sub_scenario, 'New_FA_AAC'] * EF_total_AAC

Emission_data
```

```python
Emission_data.loc['Total (with CSA)'] = Emission_data.loc['LF'] + Emission_data.loc['RC'] + Emission_data.loc['MT'] + Emission_data.loc['CSA cement']
Emission_data.loc['Total (with AAC)'] = Emission_data.loc['LF'] + Emission_data.loc['RC'] + Emission_data.loc['MT'] + Emission_data.loc['AAC']
Emission_data
```

```python
Emission_data_annual = Emission_data.copy()
Emission_data_annual[Emission_data_annual.columns[0:2]] = Emission_data_annual[Emission_data_annual.columns[0:2]] / 2
Emission_data_annual[Emission_data_annual.columns[2:]] = Emission_data_annual[Emission_data_annual.columns[2:]] / 22
```
